Avant de commencer, il est important de noter que chaque participant se trouve plusieurs fois dans le DataFrame, une fois pour chaque correspondance potentielle. Ainsi, une analyse exploratoire sur cet ensemble de données porterait sur 8378 individus avec de nombreuses répétitions.
